#### Sınıflandırma Problemlerinde Başarı Değerlendirme

```python
# öyle bir model kurayım ki müşterilerin bizi terkedip etmeyeceğini bulayım 
# Sınıf 1 Sınıf 0
# 1       1 =  1  -> true pozitif
# 0       0 =  0  -> true negatif
# 1       0  = 0  -> false pozitif
# 0       1  = 1  -> false negatif
```

* Accuracy : doğru sınıflandırma oranıdır : (TP + TN ) / ( TP + TN + FP + FN)
* Precision : pozitif sınıf (1) tahminlerinin başarı oranıdır : TP / (TP + FP)  
* Recall : Pozitif sınıfın (1) doğru tahmin edilme oranıdır 
* F1 Score : 2 * (Precision * Recall) / (Precision + Recall)

```python
def outliers_thresholds(dataframe, col_name, q1=0.05, q3=0.95):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit , up_limit
```

```python
def check_outlier(dataframe, col_name):
    low_limit, up_limit = outliers_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name]>up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False
```

```python
def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outliers_thresholds(dataframe,variable)
    dataframe.loc[(dataframe[variable]<low_limit),variable] = low_limit
    dataframe.loc[(dataframe[variable]>up_limit),variable] = up_limit
```

### Model Evalutation(Model Başarı Değerlendirme)

değerlerin numerik karşılıklarını hesaplamak için "confision matrix" fonksiyonu kullanılır

```python
def plot_confision_matrix(y,y_pred): 
    acc = round(accuracy_score(y,y_pred),2) 
    cm  = confusion_matrix(y,y_pred) 
    sns.heatmap(cm,annot=True,fmt=".0f") 
    plt.xlabel('y_pred') 
    plt.ylabel('y') 
    plt.title('Accuracy Score: {0}'.format(acc),size=10) 
    plt.show()
```

```python
plot_confision_matrix(y,y_pred)
```

```python
# recall : 156/156 + 112 
# accuracy : 1:1 0:0 tüm değerlere erişim 
# bunları genel olarak kullanımı 
# tüm değerlendirmeler 1 sınıfına göre yapılmalı 
# aritmetik ort için ikisine bakılır 0-1
print(classification_report(y,y_pred))
```
